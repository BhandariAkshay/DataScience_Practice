```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x, y) #this plots a line by joining all the points
```

```python
plt.subplot(2,3,4)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y)
plt.subplot(2,3,2)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y)
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.title('red')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*--')
plt.title('green')
```

```python
# Create figure an empty canvas
fig = plt.figure()

#adding the set of axes to the figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

#plotting on that axes
axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('x**2')
axes.set_title('Graph')
```

```python
# Create figure an empty canvas
fig = plt.figure()

#adding the set of axes to the figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

#plotting on that axes
axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('x**2')
axes.set_title('Graph')

axes1 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(y,x, 'rp')
axes1.set_xlabel('x')
axes1.set_ylabel('x**2')
axes1.set_title('Graph 1')
```

```python
#subplots with oop
fig, axes = plt.subplots()

axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('x**2')
axes.set_title('Graph')
```

```python
#subplots with oop
fig, axes = plt.subplots(2, 3)
```

```python
plt.plot(x, x+1, color = 'red', linewidth = 0.5)
plt.plot(x, x+2, color = 'red', lw = 1)
plt.plot(x, x+3, color = 'red', lw = 2)
```

```python
plt.plot(x, x+4, color = 'red', lw = 2, linestyle = ':')
plt.plot(x, x+5, color = 'red', lw = 2, linestyle = '-')
plt.plot(x, x+6, color = 'red', lw = 2, linestyle = '-.')
```

```python
fig, ax = plt.subplots(figsize = (12,3))
line,  = ax.plot(x, x+7, color = 'black', lw=1.5)
line.set_dashes([5, 10, 20, 20, 1, 5])
```

```python
plt.plot(x, x+4, color = 'red', lw = 1, marker = '+')
plt.plot(x, x+5, color = 'red', lw = 1, marker = 'o')
plt.plot(x, x+6, color = 'red', lw = 1, marker = '*', markersize = 8, markerfacecolor = 'blue')
plt.plot(x, x+7, color = 'red', lw = 1, marker = 's', markersize = 8, markerfacecolor = 'blue')
plt.plot(x, x+8, color = 'red', lw = 0.25, marker = '1')
```

```python
fig, axes = plt.subplots(1, 3, figsize = (12, 3))
axes[0].plot(x, x**2, x, x**3)
axes[0].set_title('default axes range')

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title('tight axes range')

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 40])
axes[2].set_xlim([2, 3])
axes[1].set_title('custom axes range')
```
