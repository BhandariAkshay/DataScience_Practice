```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/train.csv')
df.dtypes
```

```python
#let's look at number of missing values in each column
df.isnull().sum()
```

```python
df.drop(df[df['Embarked'].isnull()].index, axis = 0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#We will be takin care of Cabin column as it has a lot of null values
#let's create a new column has_cabin and assign it to Zero
df['has_cabin'] = 0
df
```

```python
df.loc[~(df['Cabin'].isnull()), 'has_cabin'] = 1
df
```

```python
df.drop('Cabin', axis = 1, inplace = True)
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
#as we can see here that Age still has some missing values, let's sort it out
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
# as we can see that Name, Sex, Ticket and Embarked are the columns with non integer data types
#Let's handle Embarked with categorical encoding
df.Embarked = df.Embarked.astype('category').cat.codes
```

```python
df.head()
```

```python
df.Sex = df.Sex.astype('category').cat.codes
df.head()
```

```python
df.Ticket.head()
```

```python
df.Ticket = df.Ticket.str.split().str[-1]
df.head()
```

```python
df.dtypes
```

```python
df.Ticket.astype('int')
df.dtypes
```

```python
df[df.Ticket == 'LINE']
```

```python
df.drop(df[df.Ticket == 'LINE'].index, axis = 0, inplace = True)
df[df.Ticket == 'LINE']
```

```python
df.Ticket = df.Ticket.astype('int')
```
