**%matplotlib inline**

With this inline backend, the output of plotting commands is displayed inline within frontends like the Jupyter notebook, directly below the code cell that produced it. The resulting plots will then also be stored in the notebook document.

This leads to static images

**%matplotlib notebook**

will lead to interactive plots embedded within the notebook, you can zoom and resize the figure

```python
%matplotlib notebook
df.plot()
```

**figsize(16, 4)**

Means 16 inch width by 4 inch height

**df.hist() vs. df.plot.hist()**

They do different things, df.hist() will produce a separate plot for each Series whilst df.plot.hist() will produce a stacked single plot
