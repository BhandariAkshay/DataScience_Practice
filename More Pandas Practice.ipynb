```python
df1 = pd.DataFrame({'A' : ['A1', 'A2', 'A3', 'A4'],
                   'B' : ['B1', 'B2', 'B3', 'B4'],
                   'C' : ['C1', 'C2', 'C3', 'C4'],
                   'D' : ['C1', 'C2', 'C3', 'C4']}, index = [0, 1, 2, 3])
df1
```

```python
#axis = 0, no. of rows will increase
#axis = 1, no. of columns will increase
result = pd.concat([df1,df2], axis = 1)
result
```

```python
flights = pd.read_csv('data/flights.csv')
flights
```

```python
flights.groupby('AIRLINE')['DIVERTED'].sum()
```

```python
type(flights.groupby('AIRLINE').agg({'ARR_DELAY': 'mean'}))
```

```python
type(flights.groupby('AIRLINE')['ARR_DELAY'].agg(np.mean))
```

```python
flights.groupby('AIRLINE').agg({'ARR_DELAY': 'mean'})
```

```python
flights.groupby('AIRLINE')['ARR_DELAY'].agg(np.mean)
```

```python
flights.groupby('AIRLINE')['ARR_DELAY'].mean()
```

```python
# The number of cancelled flights for every airline per day weekday
flights.groupby(['AIRLINE', 'WEEKDAY'])['CANCELLED'].sum()
```

```python
#Find the number and percentage of cancelled and diverted flights for every airline per weekday
flights.groupby(['AIRLINE', 'WEEKDAY'])['CANCELLED', 'DIVERTED'].agg(['sum', 'mean'])
```

```python
# For each origin to destination flight, find the total number of flights, 
# the number and percentage of cancelled flights and the average and variance of the airtime. 
flights.groupby(['ORG_AIR', 'DEST_AIR']).agg({'CANCELLED': ['sum', 'mean', 'size'],
                                             'AIR_TIME': ['mean', 'var']})
```
