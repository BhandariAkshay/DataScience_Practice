**DISTPLOT**

This only works on int for float dtypes

DISTPLOT is a Univariate distribution of data

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(iris['petal_width'], kde=False)
```

**Joint Plot**

Join plot allows us to match two distplot with bivariate data. 

kind parameter can be used to specify the types like scatter, hex, reg, resid, kde

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'scatter') #default is scatter
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg')
#adds linear regression fit and univariate curves(kde) same as sns.regplot()
```

```python
#this function will regress y on x possibly as a robust on polynomial equation and 
#then draw a scatterplot of the residulas same as sns.residplot()
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'resid')
```
